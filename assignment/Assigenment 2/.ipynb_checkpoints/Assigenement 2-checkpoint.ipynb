#### Bar Chart

* A bar chart is good at showing and comparing categorical data. The bar chart can even show negative values. 
* Also it is good at showing change in data over time.

* It is a good idea to avoid bar chart if the data is not categorical.
* Also it is of best practice to avoid adding too many bars in the bar chart

```python
import numpy as np
import random
import matplotlib.pyplot as plt

animals_in_zoo = ["Tiger", "Elefant", "Jaguar", "Pinguin", "Snake", "Bear", "Wolf", "Corcodile"]
numbers_of_animals = [random.randrange(3,20) for num in animals_in_zoo]

plt.bar(animals_in_zoo, numbers_of_animals)
plt.show()
```

#### Line Chart

* Line chart is good at showing data that changes over time.
* Also if the categorical data is too big to show in a bar chart.
* Showing trends is better with a line chart since you can visualize the rise and fall of trends.
* Line charts are for large dataset, and can be vague in displaying a small dataset.

```python
randome_number = random.randrange(100)
rise = [x +randome_number+ 10 for x in range(50)]
fall = [x +rise[49]-10 for x in rise]
trend = rise + fall
plt.plot(trend)
plt.show()

# In this example we try to simulate a sudden jump in the stock market.
```

#### Pie Chart / Donut Chart

* It is a good idea to use pie/donut chart when simulating the proportions and precentages of a dataset.
* Can be used to show the size of categorical data in the dataset.
* Best used when the categories is under 10.
* Dont use pie/donut if you want to show precise comparison of values.


```python
plt.pie(numbers_of_animals, labels=animals_in_zoo)
plt.show()
```

#### Stacked Area Chart

* Is good to use when you want to show correlation and clustering.
* Don't use area chart if you have a small dataset or there is no correlation in dataset.


```python
product_1 = [x + random.randrange(10) for x in range(10)]
product_2 = [x + random.randrange(5) for x in range(10)]
product_3 = [x + random.randrange(20) for x in range(10)]

product_1, product_2, product_3
```
