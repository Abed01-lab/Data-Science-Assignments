```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

####  1. This is where we load the two different dataset

```python
red = pd.read_excel("./winequality-red.xlsx")
red.assign(Name="type")
red["type"] = "red"
white = pd.read_excel("./winequality-white.xlsx")
white.assign(Name="type")
white["type"] = "white"
```

#### 2. With pandas.info() we can get an overview of the columns and the type of data, note we see the column type, that is because we added type as column to both dataframes 

```python
combined_df = pd.concat([red, white])
combined_df.head()
```

####  3. Here we will show the count, mean, min, max, std, quantiles for all the columns, at the bottom of alle the means we see the type of data for the column

```python
mean = combined_df.mean()
minimum = combined_df.min()
maximum = combined_df.max()
std = combined_df.std()
quantile = combined_df.quantile()

print("Numbers of rows in combined_df is : ", len(combined_df.index))


print("")
print("\n\nMean\n", mean,  "\n\nMinimum\n", minimum, "\n\nMaximum\n", maximum, "\n\nStd\n", std, "\n\nQuantile\n", quantile)
```

#### 4. Here we will show some diagrams that can answer these question:

 * Can you tell which type of wine has higher average quality? 
 * Which type of wine has higher average level of alcohol? 
 * Which one has higher average quantity of residual sugar? 

```python
red_df = combined_df.loc[combined_df['type'] == "red"]
white_df = combined_df.loc[combined_df['type'] == "white"]

red_avg_quality = red_df["quality"].mean()
white_avg_quality = white_df["quality"].mean()
fig1, ax1 = plt.subplots()
ax1.bar(["red", "white"], [red_avg_quality, white_avg_quality])
ax1.set_title("Quality of wine")

red_avg_alcohol = red_df["alcohol"].mean()
white_avg_alcohol = white_df["alcohol"].mean()
fig2, ax2 = plt.subplots()
ax2.pie([red_avg_alcohol, white_avg_alcohol], labels=["red", "white"], autopct='%1.1f%%',shadow=True, startangle=90)
ax2.set_title("Alcohol Levels")

red_avg_sugar = red_df["residual sugar"].mean()
white_avg_sugar = white_df["residual sugar"].mean()
fig3, ax3 = plt.subplots()
ax3.barh(["red", "white"], [red_avg_sugar, white_avg_sugar])
ax3.set_title("Avg Quantity Of Residual Sugar")

plt.show()
```

##### 7. Search the data for allocating
* a. missing values
* b. cells with a value of 0 (zero)
 
* Replace these values with the average value of their column.

```python
### missing values
for column_name in combined_df.columns:
    column = combined_df[column_name].isna()
    # Get the count of Zeros in column 
    count = column.sum()
    print('Cells with missing values in column', column_name, 'is :', count)

print("-----------------------")
### Values == 0
for column_name in combined_df.columns:
    column = combined_df[column_name]
    # Get the count of Zeros in column 
    count = (column == 0).sum()
    print('Count of zeros in column', column_name, 'is :', count)
```

```python
### Adding mean value for cells with value = 0
mean = combined_df["citric acid"].mean()
replaced_zero = combined_df["citric acid"].replace(0, mean)

print("before replacing")
print(combined_df["citric acid"].head())
print("---------")
print("after replacing")
print(replaced_zero.head())
```

##### 8. Explore the feature residual sugar. 
* Is there any outlier (a value much different from the rest)?
        - Answer: Yes there is any value over 30 in residual sugar is a outlier
* On which row is it found? Remove that row.
        - Answer: It was found on row 1653, 1663, 2781. Rows removed

```python
fig, ax = plt.subplots()
ax.scatter(combined_df["residual sugar"], combined_df["density"])
plt.show()

outliers = combined_df[combined_df["residual sugar"] > 30].index
print(outliers)
print("Before removing rows:", len(combined_df))
combined_df.drop(outliers, inplace=True)
print("After removing rows", len(combined_df))
```

##### 9. Identify the attribute with the lowest correlation to the wine quality and remove it.

```python
df_positive_corr_quality = combined_df

#df_positive_corr_quality.drop("type", 1, inplace=True)
df_positive_corr_quality.head(10)
df_positive_corr_quality["citric acid"]

for column in df_positive_corr_quality.columns:
    
    if column == "type":
        continue
    
    if df_positive_corr_quality.corrwith(df_positive_corr_quality[column])[column] < 0:
        df_positive_corr_quality.drop(column, 1, inplace=True)

df_positive_corr_quality.head(10)
```

##### 10. Finally, transform categorical data into numeric and print out the start and the end of the preprocessed data frame.

```python
import seaborn as sns

#Use the combined dataframe.
preprocessed_df = combined_df

# 1. missing values to mean value of the column.
preprocessed_df["citric acid"].replace(0, mean, inplace=True)

# 2. remove outlier.
outliers = preprocessed_df[preprocessed_df["residual sugar"] > 30].index
preprocessed_df.drop(outliers, inplace=True)

# 3. remove columns with lowest amount of correlation to wine quality.
for column_name in preprocessed_df.columns:
    
    if column_name == "type": continue
    if preprocessed_df.corrwith(preprocessed_df[column_name])[column_name] < 0:
        preprocessed_df.drop(column_name, 1, inplace=True)

# take whole combined df and transform categorial data (type of wine) to numeric values.
preprocessed_df["type"] = pd.factorize(preprocessed_df["type"])[0]

# print the start and the end of the processed dataframe
preprocessed_df
```
